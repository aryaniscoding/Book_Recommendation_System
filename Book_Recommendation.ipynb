# POPULARITY BASED RECOMMENDATION SYSTEM

```python
# TOP 50 BOOKS HIGHEST AVERAGE RATINGS BOOK , AFTER MINIMUM 250 RATINGS
```

```python
# merge books and ratings on isbn
ratings_with_name = ratings.merge(books,on = 'ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'},inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# filter the books which do not have num_ratings > 250
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending = False).head(50)
```

```python
popular_df.merge(books,on = 'Book-Title').drop_duplicates('Book-Title').shape
```

```python
popular_df = popular_df.merge(books,on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

# COLLABARATIVE RECOMMENDATION SYSTEM


```python
# we only value users who have rated more than 200 books and 
# we consider only those books which have more than 50 ratings
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

# finding the eucledian distance of all these dimensional points with each other for finding similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[123]
```

```python
# recommend funtion which returns top 5 suggestions for him
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```
